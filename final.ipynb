```python
from matplotlib import pyplot as plt
```

```python
salesdata=pd.read_csv('SalesData.csv')
salesdata
```

```python
salesdata.dropna(axis=0,inplace=True)
salesdata
```

```python
salesdata.drop_duplicates(keep='first',inplace=True)
salesdata
```

```python
salesdata.info()
```

```python
salesdata.head(5)
```

```python
sales_region=salesdata.groupby('Region')[['Sales2015','Sales2016']].sum()
sales_region
```

```python
sales_region.plot(kind='bar',figsize=(7,7),grid=True)
plt.legend()
plt.xticks(rotation=0,horizontalalignment='center')
plt.title('Compare Sales by region for 2016 with 2015 using bar chart')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

```python
salesdata_pie=salesdata.groupby('Region')[['Sales2016']].sum()
salesdata_pie
```

```python
salesdata_pie.plot(kind='pie',subplots=True,labels=['Central','East','West'],colors=['r','g','b'],autopct="%1.0f%%")
plt.legend()
plt.title('Pie chart of Sales 2016')
plt.show()
```

```python
sales_region_tier=salesdata.groupby(['Region','Tier'])['Sales2015','Sales2016'].sum()
sales_region_tier
```

```python
sales_region_tier.plot(kind='bar',figsize=(17,17))
plt.xlabel=('Tier')
plt.ylabel=('Sales')
plt.title('Compare sales of 2015 and 2016 with Region and Tiers ')
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

```python
sales_state = salesdata.groupby(['Region',"State"])['Sales2015','Sales2016'].sum()
sales_state
```

```python
sales_east=sales_state.loc['East']
sales_east
```

```python
sales_east.plot(kind="bar",figsize=(12,8),width=0.8)
plt.title("Sales comparison between 2015 and 2016 for East Region")
plt.show()
```

```python
month=salesdata['Month']
```

```python
quarter=[]

for x in month:
    if x in ['Jan','Feb','Mar']:
        quarter.append('Q1')
    elif x in ['Apr','May','Jun']:
        quarter.append('Q2')
    elif x in ['Jul','Aug','Sep']:
        quarter.append('Q3')
    else:
        quarter.append('Q4')
```

```python
quarter
```

```python
salesdata['Qtr']=pd.Series(quarter)
```

```python
salesdata.head(5)
```

```python
sales_qtr=salesdata.groupby('Qtr')['Sales2015','Sales2016'].sum()
```

```python
sales_qtr
```

```python
sales_qtr.plot(kind='bar')
plt.title('Compare Qtr wise sales in 2015 and 2016 in a bar plot ')
plt.show()
```

```python
salesdata.head()
```

```python
pivot_qtr=salesdata.pivot_table(index='Qtr',columns='Tier',values='Sales2016')
pivot_qtr
```

```python
plt.pie(x=pivot_qtr.loc["Q1",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
plt.pie(x=pivot_qtr.loc["Q2",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
plt.pie(x=pivot_qtr.loc["Q3",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```
